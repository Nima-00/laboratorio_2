### 2. Creare un Grafico

````python
import matplotlib.pyplot as plt
import random

x = [i for i in range(10)]
y_1 = [random.uniform(-1, 1) for i in range(len(x))]
y_2 = [random.uniform(-1, 1) for i in range(len(x))]

plt.figure(figsize=(6, 4))
plt.plot(x, y_1, label="y_1", color="red", linewidth=1.5, linestyle="--", marker="o")
plt.plot(x, y_2, label="y_2", color="blue", linewidth=2.0, linestyle="dotted", marker="*") 
plt.title("Grafico")
plt.xlabel("Asse x")
plt.ylabel("Asse y")
plt.legend()
plt.grid()
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import random

indeces = np.arange(10)
width = 0.3

x = [i for i in range(10)]
y_1 = [random.uniform(-1, 1) for i in range(len(x))]
y_2 = [random.uniform(-1, 1) for i in range(len(x))]

plt.figure(figsize=(6, 4))
plt.style.use("seaborn-v0_8")
plt.bar(indeces, y_1, label="y_1", width=width)
plt.bar(indeces+width, y_2, label="y_2", width=width) 

plt.title("Grafico")
plt.xlabel("Asse x")
plt.ylabel("Asse y")
plt.legend()
plt.xticks(indeces+width/2, x)
plt.show()
````

### 5. Barre orizzontali

````python
import matplotlib.pyplot as plt
import numpy as np
import random

indeces = np.arange(10)
width = 0.3

x = [i for i in range(10)]
y_1 = [random.uniform(-1, 1) for i in range(len(x))]
y_2 = [random.uniform(-1, 1) for i in range(len(x))]

plt.figure(figsize=(6, 4))
plt.style.use("seaborn-v0_8")
plt.barh(indeces, y_1, label="y_1")

plt.title("Grafico")
plt.xlabel("Asse x")
plt.ylabel("Asse y")
plt.legend()
plt.yticks(indeces)
plt.show()
````

### 6. Grafico a torta

````python
plt.style.use("ggplot")

slices = [80, 20]
labels = ["sì", "no"]
explode = [0, 0.1]
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': '#000000'}, startangle=180, explode=explode, autopct='%1.1f%%')

plt.title("Grafico a torta")
plt.show()
````

### 7. Grafico a linee

````python
import matplotlib.pyplot as plt
import random

x = [i for i in range(10)]
y = [random.uniform(-1, 1) for i in range(len(x))]
h = 0.4

plt.style.use("ggplot")
plt.plot(x, y)
plt.fill_between(x, y, h, alpha=0.5)

plt.title("Grafico")
plt.show()
````

### 8. Istogrammi

````python
import matplotlib.pyplot as plt
import random

x = [random.randrange(0, 10) for i in range(50)]

plt.style.use("ggplot")
plt.hist(x, bins= 10, color="grey", edgecolor="black")
plt.axvline(5, color="red", linewidth=3)
plt.title("Nomde del grafico")
plt.show()
````

### 9. Scatter plot

````python
import matplotlib.pyplot as plt
import random

x = [random.uniform(-1, 1) for i in range(5000)]
y = [random.uniform(-1, 1) for i in range(5000)]
colors = ["red" if x[i]**2+y[i]**2 < 1 else "blue" for i in range(len(x))]

plt.style.use("ggplot")
plt.figure(figsize=(5, 5))

plt.scatter(x, y, s=10, c=colors)
# cbar = plt.colorbar()
# cbar.set_label("Intensità")

plt.title("Titolo")
plt.show()
````
