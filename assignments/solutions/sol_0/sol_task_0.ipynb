### Esercizio 3: Tipo `float`

*Considera tre float e calcola la media e l'errore assoluto medio arrotondando alla seconda cifra decimale.*
> Memo: Utilizza la funzione `round`

````python
a = 1.0
b = 2.0
c = 3.0

m = (a+b+c)/3
mae = 0
for v in [a, b, c]:
    e = m - v
    if e < 0:           # Alternativa utilizzare abs()
        e *= -1
    mae += e
mae = mae/3
print(f"Valori: {a}, {b}, {c}")
print(f"Media =", round(m, 2))
print(f"MAE =", round(mae, 2))
````

### Esercizio 8: `while`

*Chiedi all'utente un numero intero. Dividilo per due finchè è possibile e stampa il risultato. Es: 8 -> 1, 12 -> 3.*

````python
a = int(input("Inserisci un numero intero: "))
while a % 2 == 0:
    a = a // 2
print(a)
````

*Chiedi all'utente un numero intero. Stampa il più piccolo divisore maggiore di 1.*
> Vincolo: utilizza `break`

````python
a = int(input("Inserisci un numero intero: "))
n = 2
while n <= a:
    if a % n == 0:
        break
    n += 1
print(n)
````

*Chiedi all'utente un numero intero. Stampa tutti i suoi divisori.*

> Vincolo: utilizza `continue`

````python
a = int(input("Inserisci un numero intero: "))
n = 1
while n <= a:
    if a % n != 0:
        n += 1
        continue
    print(n)
    n += 1
    
````

### Esercizio 9: `random`

Calcola la probabilità che prendendo due numeri x, y a caso nell'intervallo [-1, 1] valga x^2 + y^2 < 1.

> Memo: utilizza la funzione `random.uniform(-1, 1)` per prendere un numero casuale nell'intervallo [-1, 1].

```python
import random

a = random.uniform(-1, 1)       # Genera un numero causale nell'intervallo [-1, 1]
print(a)
```

````python
import random

tentativi = 1000000
casi_favorevoli = 0
for i in range(tentativi):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    if (x**2 + y**2) < 1:
        casi_favorevoli += 1

print(casi_favorevoli/tentativi*4)
````
