### Esercizio 1: Leggere un file e plottare i dati letti

```python
temperature_massime = {}
temperature_minime = {}
with open("temperature.txt", "r") as f:
    for line in f:
        line = line.strip()
        terms = line.split(";")
        temperature_massime[terms[0]] = [float(terms[i].replace(",", ".")) for i in range(1, len(terms)) if i % 2 == 0]
        temperature_minime[terms[0]] = [float(terms[i].replace(",", ".")) for i in range(1, len(terms)) if i % 2 == 1]
print("Temperature massime:")
print(temperature_massime)
print("Temperature minime:")
print(temperature_minime)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
mesi = ["Gen", "Feb", "Mar", "Apr", "Mag", "Giu", "Lug", "Ago", "Set", "Ott", "Nov", "Dic"]
plt.plot(mesi, temperature_massime["2024"], label="Massime 2024", marker="o")
plt.plot(mesi, temperature_minime["2024"], label="Minime 2024", marker="o")
plt.title("Temperature massime e minime del 2024")
plt.xlabel("Mese")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid()
plt.show()
```
