### Esercizio 1: Leggere un file e plottare i dati letti

```python
temperature_massime = {}
temperature_minime = {}
with open("temperature.txt", "r") as f:
    for line in f:
        line = line.strip()
        terms = line.split(";")
        temperature_massime[terms[0]] = [float(terms[i].replace(",", ".")) for i in range(1, len(terms)) if i % 2 == 0]
        temperature_minime[terms[0]] = [float(terms[i].replace(",", ".")) for i in range(1, len(terms)) if i % 2 == 1]
print("Temperature massime:")
print(temperature_massime)
print("Temperature minime:")
print(temperature_minime)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
mesi = ["Gen", "Feb", "Mar", "Apr", "Mag", "Giu", "Lug", "Ago", "Set", "Ott", "Nov", "Dic"]
plt.plot(mesi, temperature_massime["2024"], label="Massime 2024", marker="o")
plt.plot(mesi, temperature_minime["2024"], label="Minime 2024", marker="o")
plt.title("Temperature massime e minime del 2024")
plt.xlabel("Mese")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid()
plt.show()
```

### Esercizio 2: Popolazione per Regione

```python
popolazione = {}
with open("regioni.txt", "r") as f:
    for i, line in enumerate(f):
        if i == 0:
            continue
        line = line.strip()
        terms = line.split(";")
        popolazione[terms[0]] = int(terms[2])
print("Popolazione:")
print(popolazione)
```

```python
plt.figure(figsize=(14, 8))
plt.barh(popolazione.keys(), popolazione.values(), color="skyblue")
plt.title("Popolazione per regione")
plt.xlabel("Popolazione")
plt.ylabel("Regione")
plt.grid(axis="x", linestyle="--", alpha=0.5)
# for i, v in enumerate(popolazione.values()):
#     plt.text(v + 50000, i, f"{v:,}", va="center")
plt.tight_layout()
plt.show()
```

### Esercizio 3: Plottare histogramma errore

```python
# Legge dei grandi numeri
import random
def lanci_monetina(n):
    return sum(random.choice([0, 1]) for _ in range(n))
n_lanci = 1000
risultati = [lanci_monetina(n_lanci) for _ in range(1000)]
plt.hist(risultati, bins=15, color="skyblue", edgecolor="black")
plt.title(f"Distribuzione dei risultati di {n_lanci} lanci di una monetina")
plt.xlabel("Numero di teste")
plt.ylabel("Frequenza")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

### Esercizio 4: plottare una successione 

```python
y = [(1+1/i)**i for i in range(1, 101)]
plt.plot(y, marker="o")
plt.title("Approssimazione di e")
plt.xlim(0, 101)
plt.ylim(0, 4)
plt.xlabel("n")
plt.ylabel("(1 + 1/n)^n")
plt.grid()
plt.show()
```

### Esercizio 5: scatter plot

```python
import matplotlib.pyplot as plt
import random

x = [random.uniform(-1, 1) for i in range(5000)]
y = [random.uniform(-1, 1) for i in range(5000)]
colors = ["red" if x[i]**2+y[i]**2 < 1 else "blue" for i in range(len(x))]

plt.figure(figsize=(5, 5))
plt.scatter(x, y, s=10, c=colors)

plt.title("Titolo")
plt.show()
```
